```python
def reduce_categories(column, threshold=1):
    proportions = column.value_counts(normalize=True)
    small_values = set(proportions[proportions < (threshold / 100)].index)
    return column.map(lambda x: '-1' if x in small_values else x)
```

```python
# Define bin size
lat_bin_size = 0.01  # ~1.1 km
lon_bin_size = 0.01

# Create binned features
location['lat_bin'] = round((location['Latitude'] // lat_bin_size) * lat_bin_size,2)
location['lon_bin'] = round((location['Longitude'] // lon_bin_size) * lon_bin_size,2)

# Optional: Combine into one feature
location['location_bin'] = location['lat_bin'].astype(str) + '_' + location['lon_bin'].astype(str)

print(location)
```
